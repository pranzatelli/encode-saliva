```python
import numpy as np
import pandas as pd
import keras
from keras import layers
```

```python
msgG = pd.read_csv('msg_gene.csv',index_col=0); msgG = msgG.divide(msgG.sum()) * 1000000
msgG = msgG.T; maxs = msgG.max(); mins = msgG.min()
X = ((msgG - mins) / (maxs - mins)).fillna(0).values
```

```python
loss, corr = [],[]
for i in [4,5,6,7,8]:
    iloss, icorr = [],[]
    for j in range(5):
        inp = keras.Input(shape=(29058,))
        id1 = layers.Dense(50)(inp)
        id2 = layers.Dense(50)(id1)
        lv = layers.Dense(i, name="latent_vector")(id2)
        od1 = layers.Dense(50)(lv)
        od2 = layers.Dense(50)(od1)
        out = layers.Dense(29058)(od2)
        model = keras.Model(inp,out)
        model.compile(optimizer='AdamW', loss='mse')
        history = model.fit(X,X, batch_size=5, epochs = 50)
        iloss.append(history.history['loss'][-1])
        predG = model.predict(X)
        icorr.append(np.corrcoef(X.flatten(),predG.flatten())[0][1])
    loss.append(iloss); corr.append(icorr)
```

```python
for i,j in enumerate([4,5,6,7,8]):
    print(str(j)+" latent variables")
    print(np.mean(corr[i]))
```
